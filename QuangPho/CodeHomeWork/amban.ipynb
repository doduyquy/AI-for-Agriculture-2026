```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
def imread_unicode(path):
    data = np.fromfile(path, dtype=np.uint8)
    img = cv2.imdecode(data, cv2.IMREAD_COLOR)
    return img

img = imread_unicode(
    r'D:\Kaggle\Quang Ph La m\CodeHomeWork\anh-chan-dung-nu-duom-buon.jpg'
)
B = img[:, :, 0] 
G = img[:, :, 1]
R = img[:, :, 2]

gray_math = 0.299 * R + 0.587 * G + 0.114 * B
gray_math = gray_math.astype(np.uint8)


plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title('nh xm')
plt.imshow(gray_math, cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_am_ban = 255 - gray_math
plt.subplot(1, 2, 1)
plt.title('nh m bn')
plt.imshow(img_am_ban, cmap='gray')
plt.axis('off')
```
