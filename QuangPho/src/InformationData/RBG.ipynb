```python
import cv2
import numpy as np

def inspect_rgb_basic(img):
    print("=== RGB IMAGE BASIC INFO ===")
    print("Shape (H, W, C):", img.shape)
    print("Dtype:", img.dtype)
    print("Min:", img.min())
    print("Max:", img.max())
    print("Mean:", img.mean())
    print("Std:", img.std())

import cv2
import numpy as np

def imread_unicode(path):
    data = np.fromfile(path, dtype=np.uint8)
    img = cv2.imdecode(data, cv2.IMREAD_COLOR)
    return img

img = imread_unicode(
    r'D:\Kaggle\Quang Phổ Lúa mì\Data\train\RGB\Health_hyper_1.png'
)

if img is None:
    raise ValueError("Không đọc được ảnh dù file tồn tại")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
inspect_rgb_basic(img)

```

```python
def inspect_rgb_channels(img):
    channels = ["Red", "Green", "Blue"]
    for i, name in enumerate(channels):
        ch = img[:, :, i]
        print(f"\n{name} channel:")
        print("  Min:", ch.min())
        print("  Max:", ch.max())
        print("  Mean:", ch.mean())
        print("  Std:", ch.std())
inspect_rgb_channels(img)
```

```python
import matplotlib.pyplot as plt

def plot_rgb_hist(img):
    colors = ["r", "g", "b"]
    labels = ["Red", "Green", "Blue"]
    
    for i in range(3):
        plt.figure()
        plt.hist(img[:, :, i].ravel(), bins=256)
        plt.title(f"{labels[i]} Histogram")
        plt.xlabel("Pixel value")
        plt.ylabel("Count")
        plt.show()
plot_rgb_hist(img)
```

```python
def show_rgb_channels(img):
    import matplotlib.pyplot as plt
    
    names = ["Red", "Green", "Blue"]
    for i in range(3):
        plt.figure()
        plt.imshow(img[:, :, i], cmap="gray")
        plt.title(names[i])
        plt.colorbar()
        plt.show()
show_rgb_channels(img)
```

```python
def imagenet_normalize(img):
    img = img / 255.0
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])
    return (img - mean) / std

img_norm = imagenet_normalize(img)

inspect_rgb_basic(img_norm)
inspect_rgb_channels(img_norm)
```
